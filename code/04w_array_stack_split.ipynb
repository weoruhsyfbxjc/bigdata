````python
import numpy as np

# 1차원 배열 생성
a = np.array([1, 2, 3])
b = np.array([4, 5, 6])

# 2차원 배열 생성 (vstack/row_stack은 1차원 배열도 2차원으로 변환하여 스택함)
c = np.array([[7, 8, 9]])

print("Using np.vstack():")
# 1차원 배열 a와 b를 수직으로 쌓기 (결과는 2차원 배열이 됨)
result_vstack_1d = np.vstack((a, b))
print(result_vstack_1d)

print("\nUsing np.row_stack():")
# 1차원 배열 a와 b를 수직으로 쌓기 (np.vstack과 동일한 동작)
result_row_stack_1d = np.row_stack((a, b))
print(result_row_stack_1d)

print("\nUsing np.vstack() with 2D array:")
# 2차원 배열 c를 수직으로 쌓기
result_vstack_2d = np.vstack((c, c))
print(result_vstack_2d)

print("\nUsing np.row_stack() with 2D array:")
# 2차원 배열 c를 수직으로 쌓기 (np.vstack과 동일한 동작)
result_row_stack_2d = np.row_stack((c, c))
print(result_row_stack_2d)

# 두 함수의 결과가 동일한지 비교
print(f"\nAre results from vstack and row_stack (1D) identical? {np.array_equal(result_vstack_1d, result_row_stack_1d)}")
print(f"Are results from vstack and row_stack (2D) identical? {np.array_equal(result_vstack_2d, result_row_stack_2d)}")
````

# `np.random.random()` vs `np.random.default_rng().random()`

NumPy에는 난수를 생성하는 두 가지 방식이 있습니다. 겉으로는 비슷해 보이지만 내부 동작 방식과 권장 사용 시나리오가 다릅니다.

---

## 1. `np.random.random()` — 레거시 방식

```python
import numpy as np

np.random.seed(42)         # 전역 시드 설정
x = np.random.random(5)    # [0, 1) 범위의 난수 5개 생성
print(x)
```

### 특징
- **전역 상태(Global State)** 기반: 모듈 수준의 단일 난수 생성기를 공유합니다.
- **알고리즘**: Mersenne Twister (MT19937)
- **시드 설정**: `np.random.seed()`로 전역 시드를 설정 — 다른 코드와 충돌 가능성 있음
- **재현성 문제**: 멀티스레드 또는 복잡한 코드에서 시드 관리가 어려움
- **하위 호환성**: NumPy 1.0부터 사용되어 온 오래된 API

> ⚠️ NumPy 공식 문서에서는 새 코드에서는 이 방식을 **권장하지 않습니다.**

---

## 2. `np.random.default_rng().random()` — 새로운 방식 (권장)

```python
import numpy as np

rng = np.random.default_rng(seed=42)  # 독립적인 Generator 객체 생성
x = rng.random(5)                     # [0, 1) 범위의 난수 5개 생성
print(x)
```

### 특징
- **독립적인 Generator 객체**: 전역 상태를 공유하지 않아 다른 코드와 간섭 없음
- **알고리즘**: PCG64 (Permuted Congruential Generator) — MT19937보다 통계적으로 우수
- **시드 설정**: 객체 생성 시 인자로 전달 → 명시적이고 안전
- **재현성 보장**: 객체 단위로 독립 관리되어 멀티스레드 환경에서도 예측 가능
- **더 풍부한 API**: `rng.integers()`, `rng.choice()`, `rng.shuffle()` 등 일관된 인터페이스

---

## 3. 핵심 차이 요약

| 항목 | `np.random.random()` | `np.random.default_rng().random()` |
|------|----------------------|-------------------------------------|
| 도입 시기 | NumPy 초기 (레거시) | NumPy 1.17+ (권장) |
| 상태 관리 | 전역 (모듈 공유) | 독립 객체 |
| 알고리즘 | Mersenne Twister | PCG64 |
| 시드 설정 | `np.random.seed()` | `default_rng(seed=...)` |
| 통계적 품질 | 보통 | 더 우수 |
| 멀티스레드 안전성 | ❌ 불안정 | ✅ 안전 |
| 권장 여부 | ❌ 레거시 | ✅ 신규 코드에 권장 |

---

## 4. 실전 비교 예시

```python
import numpy as np

# --- 레거시 방식 ---
np.random.seed(42)
a = np.random.random(3)

np.random.seed(0)          # 어딘가에서 시드 재설정 → 재현성 깨짐
b = np.random.random(3)

np.random.seed(42)
c = np.random.random(3)    # a와 동일하다고 기대하지만, 중간에 개입 가능

print("a:", a)
print("c:", c)             # a == c (이 예시에서는 같지만, 실제 프로젝트에선 보장 어려움)

# --- 새로운 방식 ---
rng1 = np.random.default_rng(42)
rng2 = np.random.default_rng(42)   # 동일 시드 → 완전히 독립적인 두 객체

print(rng1.random(3))
print(rng2.random(3))              # rng1과 항상 동일한 결과 보장
```

## 5. 결론

> **새로 작성하는 코드에서는 항상 `np.random.default_rng()`를 사용하세요.**

- 실험 재현성이 중요한 **머신러닝 / 데이터 과학** 프로젝트
- 여러 모듈이 난수를 사용하는 **대규모 코드베이스**
- **멀티스레드 / 병렬 처리** 환경

위 모든 경우에서 `default_rng()` 방식이 훨씬 안전하고 예측 가능합니다.

레거시 방식(`np.random.random()`)은 간단한 스크립트나 기존 코드 유지보수 시에만 사용하는 것이 좋습니다.
