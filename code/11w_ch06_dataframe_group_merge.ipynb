````python
from IPython.display import display_html
import inspect
import ast

def display_side_by_side(*dfs):
    titles = []

    try:
        caller_frame = inspect.currentframe().f_back
        call_line = inspect.getframeinfo(caller_frame).code_context[0].strip()

        tree = ast.parse(call_line)
        call = tree.body[0].value

        if isinstance(call, ast.Call):
            for arg in call.args:
                titles.append(ast.get_source_segment(call_line, arg))
    except:
        titles = [''] * len(dfs)

    if len(titles) < len(dfs):
        titles += [''] * (len(dfs) - len(titles))

    html_str = ''
    for df, title in zip(dfs, titles):
        html_str += f"""
        <div style="display:inline-block; vertical-align:top; margin-right:20px; text-align:center;">
            <div style="font-weight:bold; font-size:13px; margin-bottom:8px; text-align:center;">
                {title}
            </div>
            <div style="display:flex; justify-content:center;">
                {df.to_html()}
            </div>
        </div>
        """

    display_html(html_str, raw=True)
````

````python
from IPython.display import display_html
import inspect
import ast

def display_series_side_by_side(*args):
    """여러 Series를 옆으로 나란히 표시하고,
    호출 시 사용한 표현식을 제목으로 위에 출력한다.
    """
    titles = []

    try:
        caller_frame = inspect.currentframe().f_back
        call_line = inspect.getframeinfo(caller_frame).code_context[0].strip()

        tree = ast.parse(call_line)
        call = tree.body[0].value

        if isinstance(call, ast.Call):
            for arg in call.args:
                titles.append(ast.get_source_segment(call_line, arg))
    except:
        titles = [''] * len(args)

    if len(titles) < len(args):
        titles += [''] * (len(args) - len(titles))

    html_str = ''
    for s, title in zip(args, titles):
        table_html = s.to_frame().to_html()

        html_str += f"""
        <div style="display:inline-block; vertical-align:top; margin-right:20px; text-align:center;">
            <div style="font-weight:bold; font-size:13px; margin-bottom:8px; text-align:center;">
                {title}
            </div>
            <div style="display:flex; justify-content:center;">
                {table_html}
            </div>
        </div>
        """

    display_html(html_str, raw=True)
````

pd.merge() 함수의 how 인자는 두 데이터프레임을 어떤 기준으로 합칠지 결정합니다. 주로 데이터베이스의 조인(JOIN) 연산과 동일한 개념이라고 생각하시면 됩니다. df1과 df2를 예시로 들어 설명해 드릴게요.

df1:

   id  fruits
0   1   apple
1   2  banana
df2:

   id  price
0   1   3000
1   5    500
1. how='inner' (내부 조인, 기본값)
설명: 두 데이터프레임 모두에 존재하는 키(여기서는 id)만 결합합니다. 어느 한쪽에만 있는 키를 가진 행은 결과에서 제외됩니다.
예시 (pd.merge(df1, df2, how='inner')) 결과: id가 1인 행만 두 데이터프레임에 모두 있으므로 id=1인 행만 합쳐집니다.
   id fruits  price
0   1  apple   3000
사용 시기: 양쪽 데이터에 모두 존재하는 정보만을 보고 싶을 때 사용합니다.
2. how='left' (왼쪽 조인)
설명: 왼쪽 데이터프레임(df1)의 모든 키를 기준으로 결합합니다. 오른쪽 데이터프레임(df2)에서 매칭되는 키가 없는 경우, 해당 컬럼의 값은 NaN (Not a Number)으로 채워집니다.
예시 (pd.merge(df1, df2, how='left')) 결과: df1의 id 1과 2는 모두 유지됩니다. id=1은 df2에 있으니 매칭되지만, id=2는 df2에 없으므로 price 컬럼이 NaN이 됩니다.
   id  fruits  price
0   1   apple   3000
1   2  banana    NaN
사용 시기: 왼쪽 데이터프레임의 모든 정보를 유지하면서 오른쪽 데이터프레임의 추가 정보를 가져오고 싶을 때 사용합니다.
3. how='right' (오른쪽 조인)
설명: 오른쪽 데이터프레임(df2)의 모든 키를 기준으로 결합합니다. 왼쪽 데이터프레임(df1)에서 매칭되는 키가 없는 경우, 해당 컬럼의 값은 NaN으로 채워집니다.
예시 (pd.merge(df1, df2, how='right')) 결과: df2의 id 1과 5는 모두 유지됩니다. id=1은 df1에 있으니 매칭되지만, id=5는 df1에 없으므로 fruits 컬럼이 NaN이 됩니다.
   id  fruits  price
0   1   apple   3000
1   5     NaN    500
사용 시기: 오른쪽 데이터프레임의 모든 정보를 유지하면서 왼쪽 데이터프레임의 추가 정보를 가져오고 싶을 때 사용합니다.
4. how='outer' (완전 외부 조인)
설명: 두 데이터프레임 중 어느 한쪽에라도 존재하는 모든 키를 기준으로 결합합니다. 매칭되는 키가 없는 경우, 해당 컬럼의 값은 NaN으로 채워집니다.
예시 (pd.merge(df1, df2, how='outer')) 결과: df1의 id=2와 df2의 id=5 모두 유지되며, 매칭되지 않는 부분은 NaN으로 채워집니다.
   id  fruits  price
0   1   apple   3000
1   2  banana    NaN
2   5     NaN    500
사용 시기: 두 데이터프레임의 모든 가능한 정보를 한데 모아보고 싶을 때 사용합니다.
각 조인 방식은 데이터 결합의 목적에 따라 적절히 선택하여 사용하시면 됩니다.
